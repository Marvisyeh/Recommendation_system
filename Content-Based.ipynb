## Content-Based Recommendation

```python
import numpy as np
import pandas as pd
```

```python
#read data
df = pd.read_csv('./movies_metadata.csv')
df.head()
```

```python
#inspect all columns
df.T
```

```python
#check null
df.isna().sum()
```

```python
#fillna for used columns
df.loc[:,['title','overview']] = df.loc[:,['title','overview']].fillna('')
df.isna().sum()
```

```python
#prepare data
import re
def get_title_overviews(df):
    #合併title和overview，轉換成全小寫、合併（維度（axis=1 是一維），只取values）
    title_overviews = df.apply(lambda x:x['title'].lower() + ' ' + x['overview'].lower() ,axis=1).values

    #清洗：去除符號只留英文
    title_overviews = [re.sub(r'[^a-zA-Z ]+', ' ', title_overview) for title_overview in title_overviews]

    #split to list（切割成list，因為接下來要計算TFIDF）(二維)
    title_overviews = [title_overview.split(' ') for title_overview in title_overviews]

    return title_overviews

title_overviews = get_title_overviews(df)
title_overviews[:5]
```

```python
# #計算總共有多少不同的字，先將維度轉為一維（因為向量七萬太大，所以要設定範圍）
# np.unique(np.hstack(title_overviews)).shape
```

```python
#計算TF-IDF
from collections import Counter
from sklearn.feature_extraction.text import TfidfTransformer

max_words = 10000
def get_tfidf_matrix(title_overviews):

    #先將文字轉成向量（np.hstack()將資料轉成一維array）
    #計算各個詞在電影描述中出現的次數（輸出為dic）
    counter = Counter(np.hstack(title_overviews))

    #計算出前10000較常出現的詞，按照出現頻率來排序，並將順序當成values
    word2index = {unique_word: idx for idx, (unique_word,count) in enumerate(sorted(counter.items(), key = lambda x:-x[1])) if idx < max_words}

    #初始化count_matrix，“dtype=np.int32”設定儲存方式減少存放空間
    # np.zero([第一維：總共有多少個電影],[第二維：總共考量的字詞數]) 
    count_matrix = np.zeros([len(title_overviews), max_words], dtype=np.int32)

    for idx, title_overview in enumerate(title_overviews):
        for word in title_overview:
            #如果字有在count_matrix裡面才需要考慮
            if word in word2index:
                #一向是id位置，二向是word2index的位置
                count_matrix[idx][word2index[word]] += 1
    #計算TFIDF
    transformer = TfidfTransformer()
    tfidf_matrix = transformer.fit_transform(count_matrix)
    
    #將資料轉為array
    return tfidf_matrix.toarray()

tfidf_matrix = get_tfidf_matrix(title_overviews)
tfidf_matrix.shape
    
```

### 推薦商品

```python
def get_most_similar_items(tfidf_matrix, idx, top_n):
    
    #1d array (max_words,)
    # print(tfidf_matrix[idx])
    # 2d array (num_movies, max_words)
    # print(tfidf_matrix)

    #用矩陣乘法取代迴圈
    # (num_movies, max_words)*(max_words,1) -> (num_movies,1) -> (num_movies)
    scores = np.matmul(tfidf_matrix,tfidf_matrix[idx].reshape(-1,1)).reshape(-1)
    # print(scores)

    #計算相似分數 np.argsort排序（由小到大）->np.flip倒序
    most_similar_items = np.flip(np.argsort(scores))
    # print(most_similar_items)
    #過濾掉自己，取出前top_n筆
    most_similar_items = most_similar_items[most_similar_items != idx][:top_n]
    return most_similar_items

get_most_similar_items(tfidf_matrix,4766,10)
```

```python
# 哈利波特
df.iloc[get_most_similar_items(tfidf_matrix,4766,10)].title
```
